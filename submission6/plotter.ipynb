```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
os.listdir()
```

```python
client_losses = []
client_accuracies = []

for i in range(4):
    losses = []
    accuracies = []
    with open('client-log-'+str(i)+'.txt') as file:
        lines = [line.rstrip() for line in file]
        for line in lines:
            if "ROUND START" not in line:
                line = line.split(",")
                loss = float(line[0])
                acc = float(line[1])
                losses.append(loss)
                accuracies.append(acc)
    print(len(lines) , len(losses))
    client_losses.append(losses)
    client_accuracies.append(accuracies)
```

```python
client_test_losses = []
client_test_accuracies = []

for i in range(4):
    accuracies = []
    losses = []
    with open('client-result-'+str(i)+'.txt') as file:
        lines = [line.rstrip() for line in file]
        for line in lines:
            line = line.split(",")
            loss = float(line[0])
            acc = float(line[1])
            losses.append(loss)
            accuracies.append(acc)
    client_test_accuracies.append(accuracies)
    client_test_losses.append(losses)
```

```python
for i in range(4):
    plt.plot(client_losses[i])

plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Client 1', 'Client 2', 'Client 3', 'Client 4'])
plt.show()
```

```python
for i in range(4):
    plt.plot(client_accuracies[i])

plt.title('Training Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Client 1', 'Client 2', 'Client 3', 'Client 4'])
plt.show()
```

```python
for i in range(4):
    plt.plot(client_test_losses[i])

plt.title('Testing Loss')
plt.ylabel('Loss')
plt.xlabel('Evaluation')
plt.legend(['Client 1', 'Client 2', 'Client 3', 'Client 4'])
plt.show()
```

```python
for i in range(4):
    plt.plot(client_test_accuracies[i])

plt.title('Testing Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Evaluation')
plt.legend(['Client 1', 'Client 2', 'Client 3', 'Client 4'])
plt.show()
```

```python
server_accuracies = []
server_losses = []
with open('server-eval.txt') as file:
    lines = [line.rstrip() for line in file]
    for line in lines:
        line = line.split(",")
        loss = float(line[0])
        acc = float(line[1])
        server_losses.append(loss)
        server_accuracies.append(acc)
```
